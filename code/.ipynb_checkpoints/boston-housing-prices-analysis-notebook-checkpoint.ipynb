### Importing Libraries

```python
# Importing libraries for data wrangling and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Understanding



```python
# Reading Data from the csv file (Source URL: https://www.redfin.com/city/1826/MA/Boston)
boston_real_estate_df=pd.read_csv('boston_housing_prices_analysis.csv')
print(boston_real_estate_df.shape)
boston_real_estate_df.head()
```

```python
# Getting the count of unique values for each column to understand cardinality
boston_real_estate_df.nunique()
```

As we observe from the above numbers, we can drop some of the **non-frequent fields** like Sold Date and other irrelevant fields like MLS#, Sale Type, URL, Latitude, Longitude, etc. due to **extreme high cardinality**. We will also drop fields wiith **higher missing values** as they would not contirbute much to our evaluation.

## Data Preparation Functions
Created functions to do the following tasks:
1. **Drop unnecessary columns from a data frame:** We will be using this function to remove fields that are either non-frequent, have extreme high cardinality or higher missing values. 
2. **Impute mean for NaN values in numeric columns:** As we have numeric fields with higher fill ratio, we will be using mean imputation to fill Null values with mean.

```python
def drop_columns(df,listcol):
    '''
    INPUT:
    df - pandas dataframe that you want to clean and drop columns
    listcol - list of columns that do have statistical significance 
    
    OUTPUT:
    df - the same dataframe object but with irrelevant columns dropped
    '''
    
    df.drop(columns=listcol,axis=1,inplace=True)
    return df

def mean_imputer(df):
    '''
    INPUT:
    df - data frame that needs mean imputation calculated for its numeric variables.
    
    OUTPUT:
    df - the same data frame with mean of each column imputed for NaN values
    '''
    num_vars = df.select_dtypes(include=['float','int']).columns
    for col in num_vars:
        df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
# Identifying fields with high rate of missing values with threshold of minimum 50% missing values
columns_with_high_missing_values=list(boston_real_estate_df.columns[boston_real_estate_df.isnull().mean()>0.5])

# Creating a list of the irrelevant fields from the dataframe
columns_to_drop = ['SALE TYPE','ADDRESS','STATE OR PROVINCE','ZIP OR POSTAL CODE','URL (SEE http://www.redfin.com/buy-a-home/comparative-market-analysis FOR INFO ON PRICING)','SOURCE','FAVORITE','INTERESTED','LATITUDE','LONGITUDE','MLS#']

# Combining the 2 lists of irrelevant fields for our simplified analysis
columns_to_drop = columns_to_drop + columns_with_high_missing_values # Concatenate the 2 lists
print(len(columns_to_drop)) # This is to validate the count of fields to be eliminated. For our case it should be 15.
```

```python
# Creating a copy of the original dataframe
boston_real_estate_df_clean=boston_real_estate_df.copy()

# Dropping fields to get clean version of the dataframe
boston_real_estate_df_clean=drop_columns(boston_real_estate_df_clean,columns_to_drop)
boston_real_estate_df_clean.shape # The columns should be dropped in the clean version of the dataframe
```

```python
# Feature Engineering
boston_real_estate_df_clean['AGE IN YEARS']=2021-boston_real_estate_df['YEAR BUILT']
boston_real_estate_df_clean=drop_columns(boston_real_estate_df_clean,['YEAR BUILT'])
boston_real_estate_df_clean.head()
```

We will check for all NaN values in different fields and apply our mean imputation function

```python
# Checking for null values & mean imputing numeric variables
print(boston_real_estate_df_clean.isnull().mean())

boston_real_estate_df_clean = mean_imputer(boston_real_estate_df_clean)
boston_real_estate_df_clean.isnull().mean()
```

### Q1. Which locations are hot markets in terms of high number of listings in Boston, MA area?

```python
# Top 15 locations with respect highest number of listings 
boston_real_estate_df_numlist=boston_real_estate_df['LOCATION'].value_counts()[:10].reset_index()
boston_real_estate_df_numlist=boston_real_estate_df_numlist.rename(columns={'index':'LOCATION','LOCATION':'NUMBER OF LISTINGS'})

# Visualize top 10 most hot spot neighborhoods in terms of total number of listings
plt.figure(figsize=(18, 6))
sns.barplot(x="LOCATION", y="NUMBER OF LISTINGS", data=boston_real_estate_df_numlist)
```

As we see from the above results, **Boston, South Boston and South End** areas have the maximum number of listings in Boston city neighborhoods as per this data.
These areas are generally close to downtown Boston city and more number of listings justify the popularity of the Boston city's neighborhood.

### Q2. Which are the most expensive neighborhoods in Boston, MA area?

```python
# Identifying 10 most pricey locations for purchasing homes in Boston
boston_top10_pricey = boston_real_estate_df_clean.groupby(['LOCATION'])['PRICE'].mean().reset_index().sort_values('PRICE',ascending=False)[:10]
# boston_top10_pricey['PRICE'] = (boston_top10_pricey['PRICE']/1000000).apply(lambda x:'${:,.2f}MM'.format(x))
boston_top10_pricey = boston_top10_pricey.rename(columns={'LOCATION':'LOCATION','PRICE':'AVERAGE PRICE IN ($) MILLIONS'})

# Visualize top 10 most expensive locations in terms of average price in $ millions
plt.figure(figsize=(18,6))
sns.barplot(x="LOCATION", y="AVERAGE PRICE IN ($) MILLIONS", data=boston_top10_pricey)
```

As we see from the above image, **Beacon Hill, Harvard Square and Midtown** areas have the highest average housing prices in Boston city localities as per this data. 

Places like Harvard Square, Waterfront, etc. are close to downtown Boston city area or college areas.
Overall, from the above two lists of neighborhoods we have the following observations:
1. Some of the neighborhoods that are in popular listings do also fall into the list of the most expensive localities in Boston. Eg: Beacon Hill, Back Bay, etc.
2. Most of the popular locations by count of listings are not a part of top pricey locations list in Boston area.

```python
# Understanding correlations 
sns.heatmap(boston_real_estate_df_clean.corr(),annot=True,fmt=".2f");
```

### Q3. How do housing prices vary across property types?

```python
# Data frame representing average price per property type
boston_real_estate_df_prop_price = boston_real_estate_df_clean.groupby(['PROPERTY TYPE'])['PRICE'].mean().reset_index().sort_values('PRICE',ascending=False)

# Data frame representing total count per property type
boston_real_estate_df_prop_count = boston_real_estate_df_clean['PROPERTY TYPE'].value_counts().reset_index()
boston_real_estate_df_prop_count = boston_real_estate_df_prop_count.rename(columns={"index":"PROPERTY TYPE","PROPERTY TYPE":"COUNT"})

# Merged data frame with details for each property type
boston_real_estate_df_prop_merged = pd.merge(boston_real_estate_df_prop_price, boston_real_estate_df_prop_count, on="PROPERTY TYPE")
print("SUMMARY OF PROPERTY TYPES\n")
print(boston_real_estate_df_prop_merged)
```

From the above summary of average pricing per property type, most of the listings belong to **Condo/ Co-ops**. 
However, **the average prices for Condos/ Co-ops are relatively lower** compared to other family residential listings.

### Q4. What type of impact do parameters like number of baths, beds or square feet of the plot have on the housing prices?

2. **Condos:** Pricing is variedly distributed for Condos with respect to number of beds. Condos generally do not have higher number of beds relatively.
3. **Multi-Family:** Pricing is mostly on the lower end but have high range of the number of rooms.
